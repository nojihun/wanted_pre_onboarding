```python
import numpy as np
def spli(x):
    if x == None:
        return None
    else:
        tag = x.split("|")
        return tag
```

```python
month = [g for n, g in df4.set_index('on_trending_date').groupby(pd.Grouper(freq='M'))]
month_3 = pd.DataFrame(month[0])
month_4 = pd.DataFrame(month[1])
month_5 = pd.DataFrame(month[2])
month_6 = pd.DataFrame(month[3])
month_7 = pd.DataFrame(month[4])
```

```python
def tag(x):
    from collections import Counter

    tag = []
    for i in x['tags']:
        if i != None:
            for z in i:
                z = z.replace(" ","")
                z= z.replace("#","")
                z= z.replace("-","")
                tag.append(z.lower())
    return Counter(tag)
```

```python
tag(month_3).keys()
```

```python
tag(month_4).most_common(10)
```

```python
x = []
for i in zip(*tag(month_4).most_common(10)):
    print('i:', i)
    x.append(i)
```

```python
plt.scatter(x[0], x[1])
plt.show()
```

```python
count = 1
for i in month:
    x = []
    for z in zip(*tag(i).most_common(10)):
        x.append(z)
    
    plt.figure(figsize = (10,15))
    plt.subplot(len(month),1,count)
    plt.scatter(x[0], x[1])
    plt.title(count+2)
    plt.show()
    count+=1
```

```python
df['rate'] = (df['off_likes']- df['on_likes'])/(df['off_dislikes']-df['on_dislikes'])
```

```python
df['like/dislike'] = df['on_likes']/df['on_dislikes']
df['off_like/dislike'] = df['off_likes']/df['off_dislikes']
df['commnet_change'] = df['off_comments'] -df['on_comments'] 
df['like/dislike'] = df['on_likes']/df['on_dislikes']
```

```python
def chan_t(x):
    import re
    x = x.replace("PT","")
    b= re.split('[a-zA-Z]', x)
    b=list(filter(None, b))
    c = re.split('[0-9]', x)
    c=list(filter(None, c))
    for i in range(len(c)):
        if c[i] == 'H':
            c[i] = 3600
        elif c[i] == 'M':
            c[i] = 60
        elif c[i] == 'S':
            c[i] = 1
    result=[int(b[i])*c[i] for i in range(len(b))]
    return sum(result)
    
```

```python
df['duration']=df['duration'].apply(chan_t)
```
